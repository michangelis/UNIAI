```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set to display all columns
pd.set_option('display.max_columns', None)
ontime = pd.read_csv('data/ontime.csv')
ontime
```

```python
airport = pd.read_csv('data/airport_id.csv')
airport
```

```python
cancel_codes_df = pd.read_csv('data/cancel.csv')
cancel_codes_df
```

```python
airline = pd.read_csv('data/airline_id.csv')
airline
```

```python
chicago_airport_codes = airport[airport['description'].str.contains("Chicago")]['code']

# Filter the original dataset for flights that either arrive at or depart from Chicago
chicago_flights = ontime[ontime['originairportid'].isin(chicago_airport_codes) | ontime['destairportid'].isin(chicago_airport_codes)]
chicago_flights
```

```python
# Count the number of departures from Chicago airports
departure_counts = chicago_flights[chicago_flights['originairportid'].isin(chicago_airport_codes)]['originairportid'].value_counts()

# Convert these counts to DataFrames for easier plotting
departure_counts_df = departure_counts.reset_index()

departure_counts_df = departure_counts_df.merge(airport, left_on='originairportid', right_on='code', how='left')

departure_counts_df
```

```python
# Count the number of arrivals to Chicago airports
arrival_counts = chicago_flights[chicago_flights['destairportid'].isin(chicago_airport_codes)]['destairportid'].value_counts()

arrival_counts_df = arrival_counts.reset_index()

arrival_counts_df = arrival_counts_df.merge(airport, left_on='destairportid', right_on='code', how='left')

arrival_counts_df
```

```python
# Recategorize flights into 'On-time', 'Delayed', and 'Cancelled'
chicago_flights['Status'] = chicago_flights.apply(lambda row: 'Cancelled' if row['cancelled'] == 1 else ('Delayed' if row['depdel15'] == 1 else 'On-time'), axis=1)

# Get counts of each category
status_counts = chicago_flights['Status'].value_counts(normalize=True) * 100  # Convert to percentage

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette=['green', 'red', 'blue'])

# Adding the text on top of the bars
for i in range(len(status_counts)):
    plt.text(i, status_counts[i], f'{status_counts[i]:.2f}%', ha='center', va='bottom')

plt.title('Proportion of On-time vs Delayed vs Cancelled Flights')
plt.ylabel('Percentage of Flights')
plt.xlabel('Flight Status')
plt.ylim(0, 100)  # Extend y-axis to make room for text labels
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np

# Calculate z-scores
chicago_flights['z_score'] = np.abs(stats.zscore(chicago_flights['depdelayminutes']))

# Filter rows
filtered_flights = chicago_flights[chicago_flights['z_score'] < 3]

# Create a boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(y='depdelayminutes', data=filtered_flights)

# Set the title and labels
plt.title('Box plot of Departure Delay Minutes for Chicago Flights (Outliers Removed)')
plt.ylabel('Departure Delay Minutes')

# Show the plot
plt.show()
```

```python
# Departures from Chicago
chicago_departure_delays = chicago_flights[(chicago_flights['originairportid'].isin(chicago_airport_codes)) & (chicago_flights['depdel15'] == 1)]

# Arrivals in Chicago
chicago_arrival_delays = chicago_flights[(chicago_flights['destairportid'].isin(chicago_airport_codes)) & (chicago_flights['arrdel15'] == 1)]


# Group by the destination city for departures
top_delayed_departure_destinations = chicago_departure_delays['destcityname'].value_counts().nlargest(20)

# Group by the origin city for arrivals
top_delayed_arrival_origins = chicago_arrival_delays['origincityname'].value_counts().nlargest(20)

# Plotting Departures
plt.figure(figsize=(12, 10))
sns.barplot(y=top_delayed_departure_destinations.index, x=top_delayed_departure_destinations.values, palette='coolwarm', orient='h')
plt.title('Top 20 Destination Cities for Delayed Flights Departing from Chicago')
plt.xlabel('Number of Delayed Flights')
plt.ylabel('Destination City')
plt.tight_layout()
plt.show()

# Plotting Arrivals
plt.figure(figsize=(12, 10))
sns.barplot(y=top_delayed_arrival_origins.index, x=top_delayed_arrival_origins.values, palette='coolwarm', orient='h')
plt.title('Top 20 Origin Cities for Delayed Flights Arriving in Chicago')
plt.xlabel('Number of Delayed Flights')
plt.ylabel('Origin City')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df = ontime.copy()

# Calculate the time of day in hours + fraction
df['time_of_day'] = (df['crsdeptime'] // 100) + ((df['crsdeptime'] % 100) / 60.0)

# Group by 'time_of_day' and count the number of flights
time_of_day_flight_count = df.groupby('time_of_day').size().reset_index(name='count')

# Create a DataFrame for all possible hourly time slots
all_times = pd.DataFrame({'time_of_day': np.arange(0, 24, 1/60)})  # 1-minute intervals

# Merge with the flight counts (this will fill in missing times with NaN)
full_time_of_day_flight_count = pd.merge(all_times, time_of_day_flight_count, on='time_of_day', how='left')

# Fill missing values with 0 to indicate no flights at that time
full_time_of_day_flight_count['count'] = full_time_of_day_flight_count['count'].fillna(0)

# Apply a rolling mean to smooth the plot, window size can be adjusted for more or less smoothing
full_time_of_day_flight_count['smooth_count'] = full_time_of_day_flight_count['count'].rolling(window=30, min_periods=1).mean()

# Now let's plot the number of flights throughout the day with a smooth line plot
plt.figure(figsize=(14, 7))

# Use seaborn's lineplot for a smooth curve without markers, plotting the smoothed count
sns.lineplot(x='time_of_day', y='smooth_count', data=full_time_of_day_flight_count)

# Format the x-ticks to show hours
plt.xticks(ticks=np.arange(0, 25, 1), labels=[f"{int(hour)}:00" for hour in np.arange(0, 25, 1)])

plt.xlabel('Time of Day')
plt.ylabel('Number of Flights')
plt.title('Flights per Time of Day')
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
